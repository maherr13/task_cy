```python
# Loading the checkpoints we saved from at the training phase
from extractive import ExtractiveSummarizer
model = ExtractiveSummarizer.load_from_checkpoint("/content/weights/transformerextsum-private/1urmoizn/checkpoints/epoch=1-step=933.ckpt")
```

```python
# Case 1 
text_to_summarize = """This means that on the average about 9% of the words in the human abstractive summaries do not appear in the documents.
 Since extractive automatic summarizers extract all the sentences from the documents given to them for summarization,
  clearly no extractive summarizer can have Rouge-1 recall score higher than the documents themselves on any dataset,
   and, in general, the recall score will be lower since the summaries are limited to 100 words whereas the documents themselves can be arbitrarily long."""
summary = model.predict(text_to_summarize , num_summary_sentences=1)
```
